```python
from keras.layers import Input, Lambda, Dense, Flatten,Reshape
from keras.models import Model
from keras.applications.inception_resnet_v2 import InceptionResNetV2, preprocess_input
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
#Give dataset path Define the paths
train_path = '/content/drive/MyDrive/Traffic Signs Data Set/Train'
test_path = '/content/drive/MyDrive/Traffic Signs Data Set/Test'
```

```python
InceptionResNetV2_model = InceptionResNetV2(
    include_top=False,
    weights="imagenet",
    input_shape=(299, 299, 3)
)
```

```python
import os

# Get the number of classes
folders = os.listdir('/content/drive/MyDrive/Traffic Signs Data Set/Train')
print(len(folders))
```

```python
x = Flatten()(InceptionResNetV2_model.output)
prediction = Dense(len(folders), activation='softmax')(x)
model = Model(inputs=InceptionResNetV2_model.input, outputs=prediction)
model.summary()
```

```python
from keras import optimizers

sgd = optimizers.SGD(learning_rate=0.01, momentum=0.9)
model.compile(loss='categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])
```

```python
  # Data Augmentation
  train_datagen = ImageDataGenerator(
      preprocessing_function=preprocess_input,
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')
```

```python
# Data Augmentation
test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
```

```python
train_set = train_datagen.flow_from_directory('/content/drive/MyDrive/Traffic Signs Data Set/Train',
target_size = (299, 299),
batch_size = 32,
class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory('/content/drive/MyDrive/Traffic Signs Data Set/Test',
target_size = (299, 299),
batch_size = 32,
class_mode = 'categorical')
```

```python
from datetime import datetime
from keras.callbacks import ModelCheckpoint, LearningRateScheduler
from keras.callbacks import ReduceLROnPlateau

checkpoint = ModelCheckpoint(filepath='InceptionResnet.h5',
                               verbose=1, save_best_only=True)
callbacks = [checkpoint]
start = datetime.now()
model.fit_generator(
  train_set,
  validation_data=test_set,
  epochs=10,
  steps_per_epoch=5,
  validation_steps=32,
    callbacks=callbacks ,verbose=1)



duration = datetime.now() - start
print("Training completed in time: ", duration)
```

```python
from PIL import Image
import numpy as np
from keras.preprocessing import image
import os

# Get a list of all directories (classes) in the Test folder
test_folders = os.listdir('/content/drive/MyDrive/Traffic Signs Data Set/Test')

# Iterate over each directory
for folder in test_folders:
    # Get a list of all files (images) in the current directory
    image_files = os.listdir(os.path.join('/content/drive/MyDrive/Traffic Signs Data Set/Test', folder))

    # Iterate over each image in the current directory
    for image_file in image_files:
        # Load the image
        img = Image.open(os.path.join('/content/drive/MyDrive/Traffic Signs Data Set/Test', folder, image_file))
        img = img.convert('RGB')  # Convert to RGB
        img = img.resize((299, 299), Image.ANTIALIAS)
        img = np.array(img)
        X = image.img_to_array(img)
        X = np.expand_dims(X, axis=0)
        X = preprocess_input(X)

        # Predict
        preds = model.predict(X)

        # Get the predicted class index
        pred_class_index = np.argmax(preds)

        # Print the predicted class label
        print('Predicted class for', image_file, 'in', folder, 'is:', folders[pred_class_index])
```
